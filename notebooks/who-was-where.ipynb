```python
import pandas as pd
from timelink.pandas import entities_with_attribute, display_group_attributes

# show 500 rows
pd.set_option('display.max_rows', 550)

# @wikidata:Q3558780
place = ["https://www.wikidata.org/wiki/Q3558780", "https://www.wikidata.org/wiki/Q190134"]


residents = entities_with_attribute(
    entity_type='person',
    show_elements=['id','name','groupname'],
    the_type='estadia@wikidata',
    the_value=place,
    more_attributes=['jesuita-entrada'],
    db=tlnb.db,
)
display_group_attributes(
    ids=residents.index.unique(),
    entity_type='person',
    header_elements=['id','name','groupname'],
    header_attributes=['jesuita-entrada'],
    include_attributes=['estadia','chegada','partida','jesuita-votos-local','morte',
                        ],
    sort_attributes=['the_date'],
    cmap_name='Pastel1',
    db=tlnb.db)
```
